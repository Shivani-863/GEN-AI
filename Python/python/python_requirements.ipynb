#### Operators in Python

```python
'''
Operators are symbols that perform operations on values. Think of them as the “verbs” of programming - they make things happen!
You already know most operators from math class:
Calculate: +, -, *, /
Compare: >, <, ==
Combine: and, or, not
'''

# Basic math
print(10 + 3)   # 13 - Addition
print(10 - 3)   # 7  - Subtraction
print(10 * 3)   # 30 - Multiplication
print(10 / 3)   # 3.333... - Division (always gives float)

# Special operators
print(10 // 3)  # 3  - Floor division (rounds down)
print(10 % 3)   # 1  - Modulo (remainder)
print(10 ** 3)  # 1000 - Exponent (power)

result = 2 + 3 * 4      # 14 (not 20!)
result = (2 + 3) * 4    # 20 (parentheses first)

# Comparison Operators
age = 18

print(age == 18)    # True  - Equal to
print(age != 21)    # True  - Not equal to
print(age > 17)     # True  - Greater than
print(age < 20)     # True  - Less than
print(age >= 18)    # True  - Greater than or equal
print(age <= 18)    # True  - Less than or equal

# LOgical Operators
age = 25
has_license = True

# AND - both must be true
can_drive = age >= 16 and has_license
print(can_drive)  # True

# OR - at least one must be true
day = "Saturday"
is_weekend = day == "Saturday" or day == "Sunday"
print(is_weekend)  # True

# NOT - reverses the value
is_adult = age >= 18
is_child = not is_adult
print(is_child)  # False

#Truth Tables

# AND: Both must be True
print(True and True)    # True
print(True and False)   # False
print(False and False)  # False

# OR: At least one must be True  
print(True or False)    # True
print(False or False)   # False
print(True or False)    # True
print(False or True)    # True

# NOT: Flips the value
print(not True)         # False
print(not False)        # True

# Assignment Operators
# Instead of:
score1=30
score1 = score1 + 10

# Write:
score1 += 10

# Works with all operators
x = 10
x += 5    # x is now 15
x *= 2    # x is now 30
```

#### Loops in Python

#### Data Structures in Python

Think of data structures as containers:

Lists: Like a shopping list (ordered items)

Dictionaries: Like a phone book (name > number)

Tuples: Like coordinates (fixed values)

Sets: Like a bag of unique items

```python
# What are lists?
# Lists are Python’s most versatile data structure.
# They’re like containers that can hold multiple items in a specific order.

# Empty list
my_list = []

# List with items
fruits = ["apple", "banana", "orange"]

# Get items
print(fruits[0])    # "apple" (first item)
print(fruits[1])    # "banana"
print(fruits[-1])   # "orange" (last item)
print(fruits[-2])   # "banana" (second to last)

# Slicing
print(fruits[0:2])  # ["apple", "banana"]
print(fruits[1:])   # ["banana", "orange"]

numbers = [1, 2, 3, 4, 5]
mixed = ["hello", 42, True, 3.14]  # Different types OK!

# Lists use square brackets [] and items are separated by commas. 
# You can mix different data types in the same list!
# Ordered → items keep their position
# Mutable → items can be changed
# Allow duplicates

nums=[20,10,5,78,9,"Banana",100]

print(len(nums))
# Prints elements from 1st Index to 3rd Index
print(nums[2:4])


# Prints elements from 1st Index to 3rd Index stepping by 2
print(nums[2:4:2])

# Appends "Apple" at the end of list
print(nums.append("Apple"))
print(nums) 

# Inserts 100 at 1st Index
print(nums.insert(1,100))     
print(nums)

print(nums.index(100))
print(nums)

# print(nums.index(500))  This will give error since 500 isn't there in the list
# print(nums.sort())   This will give error too snce two datatypes are present
# num1=nums.sort
print(nums)

# Removes the last most element in the list
print(nums.pop())
print(nums)

# Removes the first occurence of the 78
print(nums.remove(78))
print(nums)

# Returns count of value 
print(nums.count(100))
print(nums)

# Reverses the list
print(nums.reverse())
print(nums)
# Clears the list
print(nums.clear())
print(nums)
```

```python
# What are dictionaries? 
# Dictionaries store data in key-value pairs.
# Think of them like a real dictionary where you look up a word (key) to find its definition (value).

#Store data as key : value
# Mutable (can be changed)
# Keys must be unique
# Ordered (Python 3.7+)
# Can store mixed datatypes

# Keys must be immutable
# Allowed: strings, integers, tuples
# Not allowed: lists, dictionaries

# Dictionaries use curly braces {} with key-value pairs separated by colons. Keys must be unique
# Empty dictionary
my_dict = {}

# Dictionary with data
person = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
print(person)
# Different ways to create
scores = dict(math=95, english=87, science=92)
print(scores)
print(person["age"])
print(person.get("age"))
person["job"]="Software Developer"
print(person)
# Remove items
del person["job"]              # Remove by key
print(person)
age = person.pop("age")          # Remove and return
print(person)
person.clear()                   # Remove all items
print(person)

person = {"name": "Alice", "age": 30, "city": "New York"}

# Get all keys, values, or items
print(person.keys())    # dict_keys(['name', 'age', 'city'])
print(person.values())  # dict_values(['Alice', 30, 'New York'])
print(person.items())   # dict_items([('name', 'Alice'), ...])

# Check if key exists
if "name" in person:
    print("Name found!")

# Update multiple values
person.update({"age": 31, "job": "Engineer"})
print(person)

# Dictionary of dictionaries
students = {
    "alice": {"age": 20, "grade": "A"},
    "bob": {"age": 21, "grade": "B"},
    "charlie": {"age": 19, "grade": "A"}
}
print(students["alice"])
# Access nested data
print(students["alice"]["grade"])  # "A"
```

```python
#W hat are tuples? 
# Tuples are like lists, but they can’t be changed once created. 
# They’re immutable (unchangeable) sequences.

# Empty tuple
empty = ()

# Tuple with items
point = (3, 5)
colors = ("red", "green", "blue")

# Single item tuple needs comma!
single = (42,)  # Note the comma
not_tuple = (42)  # This is just 42 in parentheses

# Without parentheses (implicit)
coordinates = 10, 20

point = (3, 5)
colors = ("red", "green", "blue")

# Get items
print(point[0])      # 3
print(colors[-1])    # "blue"

# Slicing works too
print(colors[0:2])   # ("red", "green")

# Unpack values
point = (3, 5)
x, y = point  # x = 3, y = 5

# Multiple assignment
a, b, c = 1, 2, 3  # Same as (1, 2, 3)

# Swap variables elegantly
x, y = y, x  # Swaps values!
```

```python
# What are sets?
# Sets are collections that only store unique values. They automatically remove duplicates
# Empty set (careful!)
empty_set = set()  # NOT {} - that's a dict!

# Set with values - both ways work
numbers = {1, 2, 3, 4, 5}
fruits = set(["apple", "banana", "orange"])

# From a list (removes duplicates)
scores = [90, 85, 85, 92, 90]
unique_scores = set(scores)  # {85, 90, 92}
print(unique_scores)
# Use set() for empty sets, not {}. Empty curly braces create a dictionary!

colors = {"red", "blue"}

# Add items
colors.add("green")
print(colors)  # {'red', 'blue', 'green'}

# Remove items
colors.remove("blue")    # Error if not found
colors.discard("yellow") # No error if not found

# Check membership
if "red" in colors:
    print("Red is available")
    
# Removes duplicates
names = ["Alice", "Bob", "Alice", "Charlie", "Bob"]
unique_names = list(set(names))
print(unique_names)  # ['Alice', 'Bob', 'Charlie']
```

```python
# Function with default variables
def multiply(a,b=3):
    return a*b

# Default values are assigned always at the end of function
# def multiply2(a=10,b)  will give compilation error

print(multiply(24,5))  # Default value got overrien by 5
print(multiply(24))    # Takes default value for b... Assigns 24 to a
```

```python
# Recursion

def factorial(n):
    if n==0 or n==1:
        return 1
    return factorial(n-1)*n

print(factorial(5))
print(factorial(7))
```
