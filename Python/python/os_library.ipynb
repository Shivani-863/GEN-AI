```python
# OS Library in Python
import os
# getcwd -> Get current working directory
current_directory = os.getcwd()   
print("Current Directory:", current_directory)
 # mkdir -> Create a new directory
os.mkdir('NewFolder')
name1="Love"
name="ShivOm"
path=os.path.join(os.getcwd(),name1)  # join -> Join one or more path components
print("Joined Path:", path)
print("Is Directory:", os.path.isdir(path))
print("Is File:", os.path.isfile(path))
os.mkdir(path)  # Create the joined directory - Adds just one new directory with the name "Love" in the current working directory
# listdir -> List all files and directories in the specified path
contents = os.listdir(current_directory)
print("Is Directory:", os.path.isdir(path))
print("Contents of Current Directory:", contents)
# rmdir -> Remove a directory
os.rmdir('NewFolder')  # Remove the previously created directory
os.rmdir(path)  # Remove the joined directory
```
