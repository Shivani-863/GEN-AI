```python
try:
    print(x)
except NameError:
    print("x is not defined")
except:
    print("unknnown error")
```

```python
try:
    print(x)     # if try block gives an exception then else is also not exwcuted
except NameError:
    print("x is not defined")
else:         # if try block is executed successfully then else is also executed
    print("Success")
```

```python
try:
    print(x)     # if try block gives an exception then else is also not exwcuted
except NameError:
    print("x is not defined")
finally:         # Regardless of error like even if exception is there, finally block will get executed
    print("No matter what I'll get executed")
```
