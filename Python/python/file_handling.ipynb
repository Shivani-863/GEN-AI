```python
### File Handling in Python
# 1. Opening a File
file = open('Trial.txt', 'r')  # 'r' for read mode
# 2. Reading from a File
content = file.read()
print(content)
# 3. Closing the File
file.close()
```

```python
# Using 'with' statement for better file handling
with open('Trial.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# Writing to a File 
# In w mode it will overwrite the existing content. If you want to append to the file, use 'a' mode.
with open('Trial.txt', 'w') as file:
    file.write("This is a new line of text.\n")
    file.write("This will overwrite the existing content.\n")
with open('Trial.txt', 'a') as file:
    file.write("This line will be appended to the file.\n")
    
```

```python
#  Seek and Tell
with open('Trial.txt', 'r') as file:
    print("Current position:", file.tell())  # Get current position
    content = file.read(10)  # Read first 10 characters
    print("Content read:", content)
    print("Current position after reading:", file.tell())  # Get new position
    file.seek(0)  # Move back to the beginning of the file
    print("Position after seeking to the beginning:", file.tell())  
```
