## Pandas Series Attributes

- **values** → Returns the data values.
- **index** → Returns the index labels.
- **dtype** → Returns the data type of values.
- **size** → Returns the total number of elements.
- **shape** → Returns the shape of the Series.
- **ndim** → Returns the number of dimensions.
- **name** → Returns the name of the Series.
- **is_unique** → Checks whether all values are unique.

```python
print(df.values)
print(df.index)
print(df.dtype)
print(df.size)
print(df.shape)
print(df.ndim)
print(df.name)
print(df.is_unique)
```

```python
print(pd.Series([1,2,3,4,5,5]).is_unique)
print(pd.Series([1,2,3,4,5,]).is_unique)
```

## Pandas Series Methods

### 1. head()
- Returns the first 5 rows/elements by default.
- Used to preview the top portion of the dataset.
- You can specify the number of rows to display.
  - `head(10)` → First 10 rows
  - `head(7)` → First 7 rows

### 2. tail()
- Returns the last 5 rows/elements by default.
- Used to preview the bottom portion of the dataset.
- You can specify the number of rows to display.
  - `tail(10)` → Last 10 rows
  - `tail(7)` → Last 7 rows

### 3. sample()
- Returns a random row or rows from the dataset.
- Useful for checking random records.
- Example:
  - `sample()` → 1 random row
  - `sample(5)` → 5 random rows

### value_counts()

- Returns the count of each unique value in a Series.
- Useful for frequency analysis.
- Results are sorted in descending order by default.

Example:
`series.value_counts()`

### sort_values()

- Sorts the values in a Series.
- By default, sorting is done in ascending order.
- Use `ascending=False` for descending order.

Example:
`series.sort_values()`
`series.sort_values(ascending=False)`

### sort_index()

- Sorts a Series based on its index labels.
- By default, sorting is in ascending order.
- Use `ascending=False` for descending order.

Example:
`series.sort_index()`
`series.sort_index(ascending=False)`

```python
print(df.head())
print(df.tail())
print(df.sample(3))
df=pd.Series(['a','b','c','a','d','e','e','f','a'])
print(df.value_counts())  # Returns the frequency of each element
print(df.sort_values())
print(df.sort_values(ascending=False))

s = pd.Series(
    [100, 200, 300],
    index=['c', 'a', 'b']
)

print("Original Series:")
print(s)

print("\nSorted by Index:")
print(s.sort_index())
```

### Method Chaining

- Method chaining allows multiple methods to be called in a single statement.
- The result of one method is passed to the next method.
- Makes code concise and readable.

Syntax:
`object.method1().method2().method3()`

```python
s = pd.Series([5, 2, 8, 1, 3, 2])

result = (
    s.value_counts()
     .sort_values(ascending=False)
     .head(3)
)

print(result)
```
