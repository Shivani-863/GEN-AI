```python
import pandas as pd
mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
    'passings': [3, 7, 2]
}
myvar=pd.DataFrame(mydataset)
print(myvar)
```

```python
# Pandas Series: A Pandas Series is like a column in a table.
# It is a one-dimensional array holding data of any type.

a=[1, 7, 2]
myvar=pd.Series(a,dtype='int32')
print(myvar)

# If nothing else is specified, the values are labeled with their index number. 
# First value has index 0, second value has index 1 etc.

# This label can be used to access a specified value.
print(myvar[0])

# With the index argument, you can name your own labels.
myvar=pd.Series(a, index=["val1","val2","val3"])
print(myvar)

print(myvar["val1"])
```

```python
# Dictionary as Key data
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
# myvar=pd.Series(calories, index=["day1", "day2"])
# print(myvar)

print(myvar[0:2])
print(myvar[["day1", "day2", "day3"]]) # This is called fancy indexing.
```

```python
# Describe the series - It returns the count, mean, std, min, 25%, 50%, 75%, max values of the series.
# 25% is the first quartile, 50% is the second quartile (median), and 75% is the third quartile.
print(myvar.describe())

# sum() - It returns the sum of all the values in the series.
print(myvar.sum())
# mean() - It returns the average of all the values in the series.
print(myvar.mean())
# std() - It returns the standard deviation of all the values in the series.
# Standard deviation is a measure of how spread out the values are.
# standarad deviation is the difference between the values and the mean.
print(myvar.std())
# var() - It returns the variance of all the values in the series.
# Variance is a measure of how far a set of numbers are spread out from their average
# variance is the square of the standard deviation.
print(myvar.var())
```
