### Python functionalities with Series

##### len()-> Returns the length or size of Series
##### type()-> Returns the type of Series
##### dir()-> Returns all the methods and attributes associated with the Series
##### sorted()-> Returns the sorted Series but in case of the dataframes sorting is done by column names
##### max()-> Returns the max value in the Series
##### min()-> Returns the min value in the Series

```python
import pandas as pd
df=pd.read_excel("yt_data_two_columns.csv.xlsx", index_col=0).squeeze("columns")
print(df)
```

```python
print(len(df))
print(dir(df))
print(type(df))
print(max(df))
print(min(df))
sorted_list=sorted(df)
print(sorted_list)
```

#### Type conversion from Series to other datatypes
##### list()->Series will be converted to list
##### dict()->Series will be converted to dictionary

```python
li=list(df)
print(li)
print(type(li))
```

```python
di=dict(df)
# print(di)
di
# type(di)
```

#### Membership Operators
##### in -> Checks if the value is present in Series or not. It searches in the keys or indexes of the Series

```python
print("Anne Adkins" in df)   # Checks if the index is present in the series or not. Returns True or False
print("Shivani K" in df)   # Checks if the index is present in the series or not. Returns True or False
print(100 in df.values)   # Checks if the value is present in the series or not. Returns True or False
print(1000 in df.values)   # Checks if the value is present in the series or not. Returns True or False 
```

```python
for i in df.index:
    print(i)
# Looping directly through the series will give you the values of the series. If you want to loop through the index, you can use df.index.
for i in df:
    print(i)
```

```python
# Arithmmetic operations can be performed on series. The operations are performed on the values of the series. The index is not affected by the operations. The result of the operation is a new series with the same index as the original series.
df+100 # 100 is added to each value of the series. The result is a new series with the same index as the original series.
# Orginial series is not affected by the operation. The result of the operation is a new series with the same index as the original series.
```

```python
df2=df*50
df2
```

```python
# Relational Operators on Series
df>100 # Returns a boolean series indicating which values are greater than 100 
```

```python
df.index>"Football" # Returns a boolean series indicating which index values are greater than "Football"
```

```python
# Boolean indexing on Series
df[df>400] # Returns a series with only the values greater than 100.
           # The index of original series is preserved in the new series.
           # The index of the new series is the same as the index of the original series.
```

```python
df[df>400].size # Returns the number of values greater than 100 in the series. The index of the original series is preserved in the new series.
```
